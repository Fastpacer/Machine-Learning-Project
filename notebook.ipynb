```python
df["age"]=df["age"].astype("int")
```

```python
def age_group_creation(row):
    
    
    if row["age"]<30:
        return "20-30"
    elif row["age"]<40:
        return "30-40"
    elif row["age"]<50:
        return "40-50"
    elif row["age"]<60:
        return "50-60"
    else:
        return "60-70"
    
```

```python
df["Age Ranges"]=df.apply(age_group_creation,axis=1)
```

Feature Engineering

```python
df.columns
```

```python
X=df[["age","annual Salary","net worth"]]
y=df["car purchase amount"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.19)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
import joblib 
joblib.dump(scaler,"scaler.pkl")
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
def modelresults(predictions):
    print("Mean Absolute Error of model is {}".format(mean_absolute_error(y_test,predictions)))
    print("Root Mean Squared Error of model is {}".format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
predictions_linear_regression=lr.predict(X_test)
```

```python
modelresults(predictions_linear_regression)
```

```python
df["car purchase amount"].describe()
```

```python
from sklearn.svm import SVR
```

```python
svrmodel=SVR()
```

```python
param_gridsvr={"C":[0.01,0.1,0.5],"kernel": ["linear","rbf","poly"], "degree":[2,3,4] }
```

```python
from sklearn.model_selection import GridSearchCV
gridsvr=GridSearchCV(svrmodel,param_gridsvr)
```

```python
gridsvr.fit(X_train,y_train)
```

```python
predictions_gridsvr=gridsvr.predict(X_test)
```

```python
modelresults(predictions_gridsvr)
```

```python
gridsvr.best_params_
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel=RandomForestRegressor()
param_gridrfr={"max_depth":[8,9,10,11,12],"n_estimators":[11,12,13,14,15,16,17,18]}
```

```python
gridrfr=GridSearchCV(rfrmodel,param_gridrfr)
```

```python
gridrfr.fit(X_train,y_train.values.ravel())
```

```python
predictionsrfr=gridrfr.predict(X_test)
```

```python
modelresults(predictionsrfr)
```

```python
gridrfr.best_params_
```
